```python
#driver = webdriver.Chrome(executable_path=r"C:\Users\Saurabh\Downloads\chromedriver_win32\chromedriver.exe",chrome_options=options)
driver = webdriver.Chrome(executable_path=r"C:\Users\Saurabh\Downloads\chromedriver_win32\chromedriver.exe")
driver.maximize_window()
driver.get("http://www.rpachallenge.com/")
```

```python
driver.find_element_by_css_selector('button[class="waves-effect col s12 m12 l12 btn-large uiColorButton"]').click()

for i in df.index:
    driver.find_element_by_css_selector('input[ng-reflect-name="labelFirstName"]').send_keys(df['First Name'][i])
    driver.find_element_by_css_selector('input[ng-reflect-name="labelLastName"]').send_keys(df['Last Name'][i])
    driver.find_element_by_css_selector('input[ng-reflect-name="labelCompanyName"]').send_keys(df['Company Name'][i])
    driver.find_element_by_css_selector('input[ng-reflect-name="labelRole"]').send_keys(df['Role in Company'][i])
    driver.find_element_by_css_selector('input[ng-reflect-name="labelAddress"]').send_keys(df['Address'][i])
    driver.find_element_by_css_selector('input[ng-reflect-name="labelEmail"]').send_keys(df['Email'][i])
    driver.find_element_by_css_selector('input[ng-reflect-name="labelPhone"]').send_keys(str(df['Phone Number'][i]))
    driver.find_element_by_css_selector('input[type="submit"]').click()
driver.save_screenshot("complete.png") 
```
